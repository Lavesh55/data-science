## matplotlib pie chart

```python
y=np.array([1,5,6,9,8,4])
plt.pie(y)
plt.show()
```

```python
y=np.array([5,6,9,8])
lable=np.array(['o','n','co2','f'])
plt.pie(y,labels=lable,startangle=90)
plt.show()
```

```python
y=np.array([5,6,9,8])
lable=np.array(['o','n','co2','f'])
myexplode=[0,0,0.2,0]
plt.pie(y,labels=lable,startangle=90,explode=myexplode,shadow=True)
plt.show()
```

```python
y=np.array([5,6,9,8])
lable=np.array(['o','n','co2','f'])
color=['black','hotpink','red','blue']
myexplode=[0,0,0.2,0]
plt.pie(y,labels=lable,colors=color,startangle=90,explode=myexplode,shadow=True)
plt.show()
```

```python
y=np.array([5,6,9,8])
lable=np.array(['o','n','co2','f'])
color=['black','hotpink','red','blue']
myexplode=[0,0,0.2,0]
plt.pie(y,labels=lable,colors=color,startangle=90,explode=myexplode,shadow=True)
plt.legend(title='gases')
plt.show()
```
