```python
x=[1,2,3,4,5,6]
y=[1,2,3,4,3,6]
xx=[1,2,3,4,5,5,6]
yy=[1,2,3,4,5,2,6]

xp=np.array(x)
yp=np.array(y)
xxp=np.array(xx)
yyp=np.array(yy)

plt.subplot(1,2,1)
plt.plot(xp)
plt.plot(yp)

plt.subplot(1,2,2)
plt.plot(xxp)
plt.plot(yyp)
plt.show()
```
